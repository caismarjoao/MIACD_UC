# Transformers

The Transformer model employs an encoder-decoder structure, which is commonly utilized for tasks such as machine translation, where a series of words is translated from one language into another. This architecture is comprised of two main components:
- Encoder: Converts an input sequence of tokens into a sequence of embedding vectors, often called the hidden state or context;
- Decoder: Uses the hidden states produced by the encoder to sequentially generate an output sequence of tokens, one token at a time.

![Encoder-Decoder architecture of the Transformer](arch.png)


- The input text is tokenized and transformed into token embeddings, and positional embeddings are added because the attention mechanism doesn't recognize token order, providing necessary sequential context.
- The encoder consists of multiple encoder layers, similar to stacking convolutional layers in computer vision, with the decoder similarly built from a stack of its own layers.
- The encoder's output is fed into each decoder layer, which predicts the next token and uses this output to generate subsequent tokens until an end-of-sequence (EOS) token or a maximum length is reached.


Originally designed for sequence-to-sequence tasks like machine translation, the Transformer architecture quickly evolved, with its encoder and decoder blocks adapted into standalone models. Despite numerous variations, most Transformer models fall into one of three categories.
1. Encoder-only: These models transform text into numerical representations ideal for tasks like text classification or named entity recognition, e.g., BERT, RoBERTa, DistilBERT; 
2. Decoder-only: Models like GPT autocomplete text by iteratively predicting the next likely word, given prompts such as “What is a transformer...”. In this architecture, each token's representation is based solely on preceding context, a method known as causal or autoregressive attention.
3. Encoder-decoder: These models map complex relationships between text sequences, suitable for machine translation and summarization. Along with the Transformer architecture that includes both an encoder and a decoder.

In this lab we will focus on building an Encoder to perform text classification.



## The Enconder

The transformer's encoder features multiple stacked encoder layers, as shown in the figure bellow. Each layer processes a sequence of embeddings through subsequent sublayers:
- A multi-head self-attention layer;
- A fully connected feed-forward layer that is applied to each input embedding;

![Encoder Architecture of the Transformer](encoder.png)

These sublayers incorporate skip connections and layer normalization. Lets now see the Transformers core component: the self-attention layer.




## Self-Attention
Attention allows neural networks to variably weight elements in a sequence. In text, these elements are token embeddings, like the 768-dimensional vectors used for each token in BERT. "Self-attention" means weights are calculated for all hidden states within the same set, such as those in an encoder. Conversely, recurrent models compute attention based on the relevance of each encoder hidden state to a specific decoder state during decoding.
Self-attention uses the entire sequence to compute a weighted average for each token embedding, rather than fixed embeddings. Essentially, from a sequence of token embeddings \( x_1, ..., x_n \), self-attention generates new embeddings \( x'_1, ..., x'_n \) where each \( x'_i \) is a linear combination of all \( x_j \):

$ x_i' = \sum w_{ji} x_j $


The coefficients $ w_{ji} $, known as attention weights, are normalized so that $ \sum_j w_{ji} = 1 $. Consider the word "flies," which could evoke thoughts of insects or the verb, depending on context like in "time flies like an arrow." This understanding helps create a contextualized representation of "flies" by combining token embeddings with varying weights, emphasizing relevant contexts such as "time" and "arrow." These are known as contextualized embeddings.

### Attention Head
##### Scaled dot-product attention
Several methods exist to implement self-attention, with scaled dot-product attention being the most prevalent, as introduced in the paper "Attention Is All You Need, (2017).". Four main steps are required for this mechanism:
1. Project each token embedding into three vectors called *query*, *key*, and *value*.
2. Compute attention scores by measuring the similarity between query and key vectors using the dot product, efficiently executed via matrix multiplication. Similar queries and keys yield high dot products. The results, known as attention scores, form an $ n \times n $ matrix for a sequence of $ n $ input tokens.
3. Compute attention weights by scaling dot products to normalize variance, then applying a softmax to ensure column sums equal 1. The resulting \( n \times n \) matrix contains the attention weights, $ w_{ji} $.
4. Update token embeddings by multiplying the attention weights with the value vectors $ v_1, ..., v_n $ to produce updated embeddings $ x'_i = \sum_j w_{ji} v_j $.

**Note**: The concept of query, key, and value vectors, inspired by information retrieval systems, can be explained through a simple supermarket analogy. Consider shopping for dinner ingredients: each ingredient (query) requires matching with product labels (keys) using a similarity function. A match allows you to pick the product (value) from the shelf. Unlike the direct match in shopping, self-attention weights each label's relevance, so you might collect items in varying proportions based on how closely each label matches your query.


```python
class AttentionHead(nn.Module):
    def __init__(self, embed_dim, head_dim) -> None:
        super(AttentionHead, self).__init__()
        self.q = nn.Linear(embed_dim, head_dim, bias=False)
        self.k = nn.Linear(embed_dim, head_dim, bias=False)
        self.v = nn.Linear(embed_dim, head_dim, bias=False)
    
    
    def scaled_dot_product_attention(self, query, key, value):
        dim_k = key.size(-1)
        # the 0 dimension is the batch!
        #TODO: YOUR CODE HERE
        scores = torch.matmul(query,key.transpose(1,2)) / math.sqrt(dim_k)
        weights = F.softmax(scores, dim=-1)
        attn_outputs = torch.matmul(weights, value)
        return attn_outputs


    def forward(self, hidden_state):
        attn_outputs = self.scaled_dot_product_attention(self.q(hidden_state), self.k(hidden_state), self.v(hidden_state))
        return attn_outputs
```

### Multi-headed Attention
In our example, we directly used embeddings to compute attention scores and weights, but typically, the self-attention layer applies three separate linear transformations to each embedding, generating query, key, and value vectors. These transformations project the embeddings with unique learnable parameters, enabling the layer to target different semantic features of the sequence.

Using multiple sets of linear projections, termed attention heads, proves beneficial and is depicted in multi-head attention layer in the figure above. Multiple heads are necessary because a single head's softmax typically captures just one similarity aspect. Multiple heads allow simultaneous focus on diverse elements, such as subject-verb interactions or nearby adjectives, learning these patterns from data without manual input. This concept is akin to convolutional neural networks in computer vision, where different filters detect specific features like faces or car wheels.


Using the attention head developed above we can concatenate the outputs of each one to implement the full multi-head attention layer:

```python
class MultiHeadAttention(nn.Module):
    def __init__(self, embed_dim, num_heads) -> None:
        super(MultiHeadAttention, self).__init__()
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        self.head_dim = embed_dim // self.num_heads
        self.heads = [AttentionHead(self.embed_dim, self.head_dim) for _ in range(self.num_heads)]
        self.output_linear = nn.Linear(self.embed_dim, self.embed_dim)
    
    def forward(self, hidden_state):
        x = torch.cat([h(hidden_state) for h in self.heads], dim=-1)
        x = self.output_linear(x)
        return x
```

Now that we’ve covered attention, let’s take a look at implementing the missing piece of the encoder layer: position-wise feed-forward networks.

### The Feed-Forward Layer

The encoder and decoder's feed-forward sublayer is a simple two-layer neural network that processes each embedding independently, hence it's called a position-wise feed-forward layer. In computer vision terms, this is similar to a one-dimensional convolution with a kernel size of one. Common practice suggests the hidden size of the first layer be four times the size of the embeddings, using a GELU activation function. This sublayer is crucial for the model’s capacity and is typically scaled up to enhance model performance.

```python
class FeedForward(nn.Module):
    def __init__(self, embedd_dim, dim_ff, dropout=0.2) -> None:
        super(FeedForward, self).__init__()
        self.linear_1 = nn.Linear(embedd_dim, dim_ff)
        self.linear_2 = nn.Linear(dim_ff, embedd_dim)
        self.gelu = nn.GELU()
        self.dropout = nn.Dropout(dropout)

    
    def forward(self, x):
        x = self.linear_1(x)
        x = self.gelu(x)
        x = self.linear_2(x)
        x = self.dropout(x)
        return x
```

### Adding Layer Normalization

The Transformer architecture employs layer normalization and skip connections. Layer normalization standardizes each input in the batch to zero mean and unit variance, while skip connections forward a tensor unprocessed to the next layer, where it's added to the processed tensor. Two main placements of layer normalization within encoder or decoder layers are discussed in the literature:

- Post layer normalization: used in the original Transformer paper, situates layer normalization between skip connections. This setup can be challenging to train due to potential gradient divergence. To mitigate this, a learning rate warm-up strategy is often applied, gradually increasing the learning rate from a low to a higher value during training.
- Pre layer normalization: the most common setup in literature, integrates layer normalization within the skip connections, offering more stability during training without the need for learning rate warm-up.

We'll use the second arrangement, allowing us to assemble our components as follows:

```python
class TransformerEncoderLayer(nn.Module):
    def __init__(self, embedd_dim, num_heads, dim_ff) -> None:
        super(TransformerEncoderLayer, self).__init__()
        self.layer_norm1 = nn.LayerNorm(embedd_dim)
        self.layer_norm2 = nn.LayerNorm(embedd_dim)
        self.attention = MultiHeadAttention(embedd_dim, num_heads)
        self.feed_forward = FeedForward(embedd_dim, dim_ff)
    

    def forward(self, x):
        #TODO: YOUR CODE HERE
        hidden_state = self.layer_norm1(x)
        x = x + self.attention(hidden_state)
        hidden_state_2 = self.layer_norm2(x)
        x = x + self.feed_forward(hidden_state_2)
        return x
```

We've now built our first transformer encoder layer! However, these layers don't account for token positions, as the multi-head attention layer loses this information in its weighted sums. Fortunately, we can easily add positional data with positional embeddings. Let’s explore this.

### Positional Embeddings

Positional embeddings effectively add a position-specific vector to token embeddings, enabling attention heads and feed-forward layers to use positional data in their calculations. A popular method, particularly with large pretraining datasets, is to use learnable positional patterns, similar to how token embeddings work but based on position index instead of token ID. This approach allows for learning an efficient positional encoding during pretraining.

We can create a custom Embeddings module that merges a token embedding layer, which converts input_ids to a dense state, with a positional embedding layer for position_ids. The final embedding is the sum of these two.

```python
class Embeddings(nn.Module):
    def __init__(self, vocab_size, embedd_dim, max_position_embeddings) -> None:
        super(Embeddings, self).__init__()
        self.token_embeddings = nn.Embedding(vocab_size, embedd_dim)
        self.positional_embeddings = nn.Embedding(max_position_embeddings, embedd_dim)
        self.layer_norm = nn.LayerNorm(embedd_dim, eps=1e-12)
        self.dropout = nn.Dropout()


    def forward(self, input_ids):
        seq_length = input_ids.size(1)
        position_ids = torch.arange(seq_length, dtype=torch.long).unsqueeze(0)
        token_embeddings = self.token_embeddings(input_ids)
        position_embeddings = self.positional_embeddings(position_ids)
        embeddings = token_embeddings + position_embeddings
        embeddings = self.layer_norm(embeddings)
        embeddings = self.dropout(embeddings)
        return embeddings
```

The embedding layer now produces a dense, single embedding for each token. While learnable position embeddings are straightforward and common, there are alternatives:
- Absolute positional representations: Transformer models often use static sine and cosine patterns to encode token positions, which is particularly effective with smaller data volumes.
- Relative positional representations: emphasize the importance of nearby tokens by encoding the positions relative to each other, rather than absolute positions. This approach modifies the attention mechanism itself, incorporating additional terms that account for the distances between tokens, rather than simply adding a new relative embedding layer. This allows the model to adjust embeddings based on the relative positions of tokens during attention calculations.


Now, let's put all of this together and build the full transformer encoder combining the embeddings with the encoder layers:

```python
class TransformerEncoder(nn.Module):
    def __init__(self, num_hidden_layers, embedd_dim, num_heads, dim_ff, vocab_size, max_position_embeddings) -> None:
        super(TransformerEncoder, self).__init__()
        self.embeddings = Embeddings(vocab_size, embedd_dim, max_position_embeddings)
        self.layers = nn.ModuleList([TransformerEncoderLayer(embedd_dim, num_heads, dim_ff) for _ in range(num_hidden_layers)])


    def forward(self, x):
        x = self.embeddings(x)
        for layer in self.layers:
            x = layer(x)
        return x
```

A thats it! We have just built a Trasnformer Encoder! Now lets put all of this to the test, by building adding a classification head, and performing a classification task!

### Adding a Classification Head

Transformer models typically consist of a task-independent body and a task-specific head. So far, we've built the body of the Transformer. To create a text classifier, we need to add a classification head. Although each token has a hidden state, only one prediction is necessary. Commonly, the first token's state is used for this purpose, enhanced with a dropout and a linear layer for classification. The class below extends the existing encoder to enable sequence classification:

```python
class TransfomerForSequenceClassification(nn.Module):
    def __init__(self, num_hidden_layers, embedd_dim, num_heads, dim_ff, vocab_size, max_position_embeddings, num_classes) -> None:
        super(TransfomerForSequenceClassification, self).__init__()
        self.encoder = TransformerEncoder(num_hidden_layers, embedd_dim, num_heads, dim_ff, vocab_size, max_position_embeddings)
        self.dropout = nn.Dropout()
        self.classifier = nn.Linear(embedd_dim, num_classes)


    def forward(self, x):
        x = self.encoder(x)
        x = self.dropout(x)
        x = x.mean(dim=1)
        x = self.classifier(x)
        
        return F.log_softmax(x, dim=1)
```

### Using a Transformer to classify text

Text classification has multiple uses like categorizing customer feedback or routing support tickets by language. It's likely that your email's spam filter also employs text classification to shield your inbox from unwanted junk.
Another frequent text classification task is sentiment analysis, which seeks to determine the polarity of text.

You're a data scientist tasked with creating a system to automatically detect emotions like "anger" or "joy" expressed about your company’s product on Twitter. Lets see how we can use a transformer to do such task!

### Load the Dataset 

```python
emotions = load_dataset("imdb")
emotions.set_format(type='pandas')
df_train = emotions['train'][:]

df_train.head()
```

```python
def label_int2str(row):
    return emotions['train'].features['label'].int2str(row)
print(df_train['label'])
df_train['label_name'] = df_train['label'].apply(label_int2str)
print(df_train['label_name'])
df_train['label_name'].value_counts().plot.bar()
```

### From Text to Tokens
We are going to use a pretrained tokenizer to help us convert from words to tokens

```python
emotions.reset_format()
model_ckpt = 'distilbert-base-uncased'
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
```

```python
def tokenize(batch):
    return tokenizer(batch['text'], padding=True, truncation=True)
emotions_encoded = emotions['train'].map(tokenize, batched=True, batch_size=None)
print(emotions_encoded)

print(emotions_encoded['text'][0])
print(emotions_encoded['label'][0])
print(emotions_encoded['input_ids'][0])
```

### Model Parameterisation

```python
device = torch.device("cuda" if torch.cuda.is_available() else "mps" if torch.backends.mps.is_available() else "cpu")
device='cpu'
print(device)
num_tranformer_layers = 1
embedd_dim = 768
num_heads = 1
dim_ff = embedd_dim * 4
num_classes = len(emotions['train'].features['label'].names)
epochs = 3
data_in_use = 5000
encoder_classifier = TransfomerForSequenceClassification(num_tranformer_layers, embedd_dim, num_heads,dim_ff, tokenizer.vocab_size, tokenizer.model_max_length, num_classes)

indexes = torch.randperm(len(emotions_encoded['input_ids']))
data_in_use = indexes[:data_in_use]

train_inputs_ids = torch.tensor(emotions_encoded['input_ids'])[data_in_use]
train_labels = torch.tensor(emotions_encoded['label'])[data_in_use]
print(train_inputs_ids.shape)
print(train_labels.shape)  
td = TensorDataset(train_inputs_ids, train_labels)
loader = DataLoader(td, shuffle=True, batch_size=64)
print("n_classes",num_classes)
print("train_labels", train_labels.size())
```

### Training the Model

```python


encoder_classifier = encoder_classifier.to(device)
for param in encoder_classifier.parameters():
    param.grad = torch.zeros_like(param)

encoder_classifier.train(True)

optimizer = torch.optim.Adam(encoder_classifier.parameters(), lr=0.001)
loss_function = torch.nn.CrossEntropyLoss()

# %%
for epoch in range(epochs):
    batch_number = 0
    print('EPOCH {}:'.format(epoch + 1))
    for x_train, y_train in loader:
        x_train = x_train.to(device)
        y_train = y_train.to(device)
        optimizer.zero_grad()
        y_pred = encoder_classifier(x_train)
        train_acc = torch.sum(y_train == torch.argmax(y_pred, dim=1)) / y_train.shape[0]

        loss = loss_function(y_pred, y_train)
        loss.backward()
        optimizer.step()
        batch_number += 1
        print(f'Batch: {batch_number:3d} - Loss: {loss:.3f} - TrainAcc: {train_acc:.3f}')

encoder_classifier = encoder_classifier.to('cpu')    
print("Done!")

```

You can save the trained model for later using the Pickle module

```python
import pickle

save_model_to_file = True
load_model_from_file = False

if save_model_to_file:
    # Save the trained model as a pickle string.
    saved_model = pickle.dumps(encoder_classifier)

    # Save the pickled model to a file
    with open('encoder_classifier.pkl', 'wb') as f:
        f.write(saved_model)

if load_model_from_file:
    #load the pickled model
    with open('trained/encoder_classifier.pkl', 'rb') as f:
        encoder_classifier = pickle.load(f)
```

### Measuring the performance of the Model

```python
emotions_encoded_test = emotions['test'].map(tokenize, batched=True, batch_size=None)
test_inputs_ids = torch.tensor(emotions_encoded_test['input_ids'])[data_in_use]
test_labels = torch.tensor(emotions_encoded_test['label'])[data_in_use]
print(torch.unique(test_labels))
print("test_labels", test_labels.size())
test_td = TensorDataset(test_inputs_ids, test_labels)
test_loader = DataLoader(test_td, shuffle=False, batch_size=1000)
```

```python
encoder_classifier = encoder_classifier.to(device)
with torch.no_grad():
    encoder_classifier.eval()
    full_test_acc = []
    i = 0
    for x_test, y_test in test_loader:
        x_test = x_test.to(device)
        y_test = y_test.to(device)
        y_pred = encoder_classifier(x_test)
        test_acc = torch.sum(y_test == torch.argmax(y_pred.cpu(), dim=1)) / y_test.shape[0]
        print(f'[{i:}]TestAcc: {test_acc:.3f}')
        full_test_acc.append(test_acc)
print(sum(full_test_acc) / len(full_test_acc) )
encoder_classifier = encoder_classifier.to('cpu')
```
