<h2><font color='#3498db'>Numpy</font></h2>

NumPy’s main object is the ***homogeneous*** multidimensional array. It is a table of numberrs, ***all of the same type***, indexed by a tuple of non-negative integers. In NumPy dimensions are called axes.

The point \[1,2,3\] in the 3D space has one axis with 3 elements, i.e., it has length 3. 

The array

`[[1., 0., 0.], [0., 1., 2.]]` has 2 axes. In the first axis it has a length of 2, and in the second axis it has a length of 3.

The numpy's array class is called `ndarray`. We can also use the alias `array`. The most relevante atributes of the class `ndarray`are:

- ***ndarray.ndim*** - the number of axes (dimensions) of the array.

- ***ndarray.shape*** - the dimensions of the array. This is a tuple of integers indicating the size of the array in each dimension. For a matrix with n rows and m columns, shape will be (n,m). The length of the shape tuple is therefore the number of axes, ndim.

- ***ndarray.size*** the total number of elements of the array. This is equal to the product of the elements of shape.

- ***ndarray.dtype*** - an object describing the type of the elements in the array. One can create or specify dtype’s using standard Python types. Additionally NumPy provides types of its own. numpy.int32, numpy.int16, and numpy.float64 are some examples.

- ***ndarray.itemsize*** -  the size in bytes of each element of the array. For example, an array of elements of type float64 has itemsize 8 (=64/8), while one of type complex32 has itemsize 4 (=32/8). It is equivalent to ndarray.dtype.itemsize.

- ***ndarray.data*** - the buffer containing the actual elements of the array. Normally, we won’t need to use this attribute because we will access the elements in an array using indexing facilities.

Lets see an example on how to use this methods.

```python
# arange: Return evenly spaced values within a given interval. 
# Values are generated within the half-open interval [start, stop). Start is optional
# reshape: Returns an array containing the same data with a new shape.
a = np.arange(15).reshape(3,5) 
a
```

```python
a.shape
```

```python
a.ndim
```

```python
a.dtype.name
```

```python
a.itemsize
```

```python
a.size
```

##### Common error
Creating an array passing multiple parameters, rather that using a sequence or iterable

```python
a = np.array(1, 2, 3, 4)
```

```python
a = np.array((1, 2, 3, 4))
a
```

The array method transforms the sequences of sequences objectos into n-dimensional arrays.

```python
b = np.array([(1.5, 2, 3), (4, 5, 6)])
b
```

It is also possible to define the data types at creation time

```python
c = np.array([[1, 2], [3, 4]], dtype=np.float64)
c
```

```python
c = np.array([[1.2, 2.2], [3, 4]], dtype=np.int32)
c
```

Numpy also offers functions to create arrays with initial placeholder content. These minimize the necessity of growing arrays, an expensive operation.

`zeros`creates an array full of zeros

```python
np.zeros((3, 4))
```

`ones` creates an array full of ones

```python
np.ones((3, 4))
```

`empty` creates an array whose initial content is random and depends on the state of the memory

```python
np.empty((3, 4))
```

Above we also saw the creation of an array using `arange`. However it shouldn't be used with floating point arguments, due to the finite floating point precision. When you need to create intervals that contain floating points, it is usually better to use the function `linspace` that receives as an argument the number of elements that we want, instead of the step:

```python
np.linspace(0, 2, 9) # 9 numbers evenly spaced betwen 0 a 2
```

#### Basic Operations
Arithmetic operators on arrays apply elementwise. A new array is created and filled with the result.

```python
a = np.array([20, 30, 40, 50])
b = np.arange(4)
```

```python
a + b
```

```python
a-b
```

```python
b**2
```

```python
np.sin(a)
```

```python
a < 35
```

***IMPORTANT:*** The product operator * operates *elementwise* in NumPy arrays. The matrix product can be performed using the `@` operator

```python
A = np.array([[1, 1], [0, 1]])
B = np.array([[2, 0], [3, 4]])
```

Elementwise product

```python
A*B
```

Matrix Product

```python
A@B
```

```python
A.dot(B)
```

Many unary operations, such as computing the sum of all the elements in the array, are implemented as methods of the ndarray.

```python
a = np.arange(12).reshape(3,4)
a
```

```python
a.sum()
```

```python
a.min()
```

```python
a.max()
```

```python
a.mean()
```

By default, the operations see the array as a continuous list of numbers, regardless of its shape. However, by specifying the axis parameter you can apply an operation along the specified axis of an array:

```python
# Sum of the columns of a
a.sum(axis=0)
```

```python
# Min of the rows of a
a.min(axis=1)
```

#### Indexing, Slicing and Iterating
**One-dimensional** arrays can be indexed, sliced and iterated over, much like lists and other Python sequences.

```python
a = np.arange(10)**3
a
```

```python
a[2]
```

```python
a[2:5]
```

```python
# from start to position 6, exclusive, set every 2nd element to 1000
a[:6:2] = 1000 # same as a[0:6:2] = 1000
a
```

```python
a[::-1] # reverse the list
```

**Multidimensional** arrays can have one index per axis. These indices are given in a tuple separated by commas:

```python
a = np.arange(9).reshape(3,3)
a
```

```python
a[1,1]
```

```python
a[0:3, 1]  # each row in the second column. b[:, 1] is equivalent
```

```python
a[1:3, :] # each column in the second and third row
```

When fewer indices are provided than the number of axes, the missing indices are considered complete slices `:

```python
a[-1]   # the last row. Equivalent to a[-1, :]
```

#### Stacking
Arrays can be stacked together along different axes:

```python
a = np.arange(3)
b = np.arange(3,6)
```

```python
a
```

```python
b
```

```python
np.vstack((a,b)) # vertical stack
```

```python
np.hstack((a,b)) # horizontal stack
```

#### Copies
Simple assignments make no copy of objects or their data.

```python
a = np.arange(12).reshape(3,4)
a
```

```python
b = a  # no new object is created (shallow copy)
```

```python
b is a
```

```python
id(a) == id(b)
```

```python
d = a.copy() # a new array object with new data is created (deep copy)
```

```python
d is a
```

```python
id(a) == id(d)
```

#### Linear Algebra

```python
a = np.array([[1.0, 2.0], [3.0, 4.0]])
a
```

```python
a.transpose()
```

```python
np.linalg.inv(a)
```

```python
np.eye(2) # Identity matrix
```

```python
a @ np.eye(2)
```
