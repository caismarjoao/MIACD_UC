<h2><font color='#3498db'>Pandas</font></h2>
Pandas is built on top of NumPy. It relies on numpy's multi-dimensional arrays and fast operations to provide higher level methods for manipulation and analysis. Almost every Pandas method returns a (modified) copy of the data, which allows you to chain transformations, and perform complex modifications in one line.

Pandas provides in-memory 2D table object called Dataframe. It is like an excel spreadsheet with column names and row labels.

With these data structures, Pandas is offers additional functionalities like computing columns based on other columns and plotting graphs.

The most common data structures in pandas are:
- **Series** - 1D array, similar to a column in a spreadsheet
- **DataFrame** -  2D table, similar to a spreadsheet

Pandas’ data structures can hold mixed typed values as well as labels, and their axes can have names set.


#### Series
It is  1D labeled array. Series have only one axis (axis == 0) called “index”.

```python
pd.Series([2, 5, 'SLB', np.nan], index=['Num', 'Goals', 'Teams', 'Coach'])
```

#### DataFrames
It is  2D labeled data structure, with two axes: *index* (axis 0) and *columns* (axis 1).  DataFrames can be thought of as Python dictionaries where the keys are the column labels, and the values are the column Series.

```python
pd.DataFrame({'day': [17, 30], 'month': [1, 12], 'year': [2010, 2017]})
```

#### Accessing and setting data
To retrieve data from a DataFrame you can use several methods. They can be separated by whether they return a scalar/subset of data, or if they are label/position based.

|                | Based on Labels | Based on Position |
|----------------|-----------------|-------------------|
| Scalar         | at              | iat               |
| Subset of data | loc             | iloc              |
|                |                 |                   |


```python
df_wine.at[1,'pH']
```

```python
# Returns a Series of name pH, with index 0 and 4
df_wine.loc[[0,4],'pH']
```

```python
# Returns a Series of name 4, and the pH and alcohol associated with that row
df_wine.loc[4,['pH', 'alcohol']]
```

```python
# Returns a DataFrame with rows 0 and 4, and the pH and alcohol columns
df_wine.loc[[0,4],['pH', 'alcohol']]
```

##### Boolean Queries
Boolean queries are usefull when you do not know the indexes of the data. 

To make a boolean query, you need to pass the DataFrame a True/False Series whose index aligns with the DataFrame being queried’s index. As such, for every label in the DataFrame, there must be True/False value associated to that same label in the Series. The True/False values in the Series indicate if the value associated to a certain label should be returned in the query or not.

```python
my_series = pd.Series(np.arange(6))
```

```python
filter_mask = pd.Series([False, True, False, True, False, True, True])
```

```python
my_series[filter_mask]
```

The Series can be assembled combining other boolean queries (the query “terms”) using boolean operators.
Consider de following example, where we want to filter the original wine dataframe by wines that have alcohol > 10 and have a quality > 8

```python
df_wine[
    (df_wine.alcohol > 10) 
    & (df_wine['quality'] > 8) # two ways of accessing a column
]
```

#### Combining Data Frames
Sometimes we have information scattered, and we need to combine it into one DataFrame. One way to do this is to concatenate the DataFrames. *Concatenate* DataFrames means sticking them together along an axis, regardless of the values contained within them. Pandas’ main method for concatenation is `concat`, although you can use `append` too.

`concat` receives as argument a list of DataFrames and uses rows (axis=0)  by default, which means that it sticks one DataFrame below the ones before. If columns (axis=1) is passed, then it will stick each DataFrame at the right side of the ones before. `concat`, is a method of the module pandas.

```python
df1 = pd.DataFrame({'a': [1]})
df2 = pd.DataFrame({'b': [2]})
```

```python
# 1) Error: Raises AttributeError since concat is not a DataFrameMethod
df1.concat(df2) 
```

```python
# 2) No error, adds bellow
pd.concat([df1, df2])
```

```python
# 2) No error, adds to the right
pd.concat([df1, df2], axis=1)
```

```python
# 3) No error , and equivalent to 2)
df1.append(df2)     
```

#### GroupBy
Pandas’ GroupBy is what you would expect: a set of methods to perform splitting into groups, transforming them and combining the results. It is recommended to read the Pandas documentation about [Group by: split-apply-combine](https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html)

The main method for grouping data is `groupby`. You can invoke it directly from a DataFrame object, simply by passing a list of the columns you wish to group the DataFrame by. It is possible to group by any axis, but by default, the grouping is made through index (rows).However you could group by columns.

One important thing that you need to take into account is that the `groupby` method is lazy for sake effiency, i.e., it does not perform the grouping until it is really needed.


```python
bins = [0, 3, 6, 9, np.inf]
labels = ['bad', 'good', 'very good', 'excellent']
quality_groups = pd.cut(df_wine['quality'], bins, labels=labels)
df_wine['quality_groups'] = quality_groups
```

The example above shows a method called `cut` that allows us to bin the data according to a column. In the example, we want to classify the wine by their quality. We start by defining the bins, i.e., the values that delimit the quality intervals we want. Then we define the labels, i.e., the name of each bin/interval. Thirdly we use the `cut` method to map the wine's quality to a label. finally we add the groups to our DataFrame.

Now we can use the created column to group the wines by quality.

```python
groups = df_wine.groupby(['quality_groups'])
```

“groups” is an object of type DataFrameGroupBy. It can be iterated in order to inspect or work with the different groups (each one a DataFrame object) You can also use the get_group method (in this case passing it a 2-sized tuple) to get the corresponding DataFrame group.

The first thing we would like to see about the resulting groups is the size of each one. To do this you can use the size method, which will return a multi-level Series. Each level corresponds to a grouping field, and the values correspond to the number of rows associated to the group.

```python
groups.size()
```

Given that the result is a series, one can get the percentage of each group in relathion to the whole dataset.

```python
groups.size() / len(df_wine) * 100
```

With this, it is possible to conclude that we have few bad wines :)
